## 시각화 라이브러리
* matplotlib
* seaborn

```python
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,14,100)
```

```python
y1=np.sin(x)
y2=np.sin(x)*2
y3=np.sin(x)*4

# 그래프 그리기
plt.plot(x,y1,x,y2,x,y3)
```

```python
y1=np.sin(x)
y2=np.sin(x)*2
y3=np.sin(x)*4

# 그래프 그리기
plt.plot(x,y1,label='sin(x)')
plt.plot(x,y2,label='sin(x)*2')
plt.plot(x,y1,label='sin(x)*4')
plt.legend()
plt.title('sin plot')
```

```python
z1=np.cos(x)
z2=np.cos(x)*2
z3=np.cos(x)*4

# 그래프 그리기
plt.plot(x,z1,'o', label='cos(x)')
plt.plot(x,z2,'v', label='cos(x)*2')
plt.plot(x,z3,'^', label='cos(x)*4')
plt.legend()
plt.title('cos plot')
```
